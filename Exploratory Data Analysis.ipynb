# Data Visualisation

```python
sns.pairplot(data)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(11,8))
sns.heatmap(data.corr(), annot = True)
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,8))
sns.distplot(data['Sales'],ax=axes[0,0])
sns.distplot(data['Profit'],ax=axes[0,1])
sns.distplot(data['Discount'],ax=axes[1,0] )
sns.distplot(data['Quantity'],ax=axes[1,1])
plt.show()
```

```python
fig,axes = plt.subplots(1, 2,figsize=(15,5))
data.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Sales & Profit Relation based on Sub-Category')
data.groupby('Sub-Category')['Sales','Quantity'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Sales & Quantity Relation based on Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize=(14,5))
data.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Quantity & Profit Relation based on Sub-Category')
data.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Discount & Profit Relation based on Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(15,8))
fig.suptitle("Sales with different shipping modes and Segments", fontsize=16)
sns.barplot(x = data['Ship Mode'], y = data['Sales'],ax=axes[0])
sns.barplot(x = data['Segment'], y = data['Sales'],ax=axes[1])
plt.show()
```

```python
fig,axes = plt.subplots(2,1,figsize=(18,8))
fig.suptitle("Quantity & Profit Relation based on sub category", fontsize=16)
sns.barplot(x = data['Sub-Category'], y = data['Quantity'],ax=axes[0])
sns.barplot(x = data['Sub-Category'], y = data['Profit'],ax=axes[1])
plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize=(15,8))
fig.suptitle("Profit & Sales relation in different category", fontsize=16)
sns.barplot(data['Category'], data['Profit'], ax=axes[0])
sns.barplot(data['Category'], data['Sales'], ax=axes[1])
```

```python
sns.barplot(data['Category'], data['Discount'])
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,7))
sns.countplot(data['Quantity'],hue=data['Region'])
plt.show()
```
